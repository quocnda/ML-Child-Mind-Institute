## 2.2. Lọc các hàng dữ liệu

Thống kê cơ bản các hàng dữ liệu với ngưỡng 60% và 70%

```python
# Tính tổng số giá trị bị thiếu trên mỗi hàng
row_missing_ratio = selected_dtrain.isnull().mean(axis=1)

# Số hàng thiếu > 60% dữ liệu trên tổng số hàng
rows_missing_over_60 = (row_missing_ratio > 0.6).sum()
percentage_missing_over_60 = rows_missing_over_60 / len(selected_dtrain) * 100

# Số hàng có label mà thiếu > 60% dữ liệu trên tổng số hàng có label
rows_with_label = selected_dtrain['sii'].notnull()  # Chỉ các hàng có label không thiếu
rows_missing_over_60_with_label = ((row_missing_ratio > 0.6) & rows_with_label).sum()
percentage_missing_over_60_with_label = rows_missing_over_60_with_label / rows_with_label.sum() * 100

# Số hàng thiếu > 70% dữ liệu trên tổng số hàng
rows_missing_over_70 = (row_missing_ratio > 0.7).sum()
percentage_missing_over_70 = rows_missing_over_70 / len(selected_dtrain) * 100

# Số hàng có label mà thiếu > 60% dữ liệu trên tổng số hàng có label
rows_missing_over_70_with_label = ((row_missing_ratio > 0.7) & rows_with_label).sum()
percentage_missing_over_70_with_label = rows_missing_over_70_with_label / rows_with_label.sum() * 100

# In kết quả
print(f"Số hàng thiếu > 60% dữ liệu: {rows_missing_over_60}/{len(selected_dtrain)} ({percentage_missing_over_60:.2f}%)")
print(f"Số hàng có label mà thiếu > 60% dữ liệu: {rows_missing_over_60_with_label}/{rows_with_label.sum()} ({percentage_missing_over_60_with_label:.2f}%)")

print(f"Số hàng thiếu > 70% dữ liệu: {rows_missing_over_70}/{len(selected_dtrain)} ({percentage_missing_over_70:.2f}%)")
print(f"Số hàng có label mà thiếu > 70% dữ liệu: {rows_missing_over_70_with_label}/{rows_with_label.sum()} ({percentage_missing_over_70_with_label:.2f}%)")
```

Vì số hàng có label là không nhiều, nên không thể lãng phí được. Việc chọn ngưỡng 60% để loại bớt hàng lâ hơi tiếc nên mình mạnh dạn chọn ngưỡng 70%

## 2.2. Encode cho features kiểu object

```python
# Lấy danh sách các cột kiểu 'object'
is_cat = selected_dtrain_cleaned.dtypes == 'object'
cat_cols = selected_dtrain_cleaned.columns[is_cat]

# Tạo bảng cat_table, loại bỏ cột 'id' nếu tồn tại
if 'id' in cat_cols:
    cat_table = selected_dtrain_cleaned[cat_cols].drop(columns=['id'])
else:
    cat_table = selected_dtrain_cleaned[cat_cols]

# Loại bỏ các cột không có trong `same_columns`
cat_table = cat_table[[column for column in cat_table.columns if column in same_columns]]

print("Các features dạng object / category có trong tập train&test là:")
cat_table
```

Nhận xét: toàn bộ features dạng object/category đều biểu diễn mùa, nên từ giờ mình dùng _mùa_ để nói về kiểu dữ liệu này. Như vậy data của mình có 2 kiểu dữ liệu chính: _numeric_ và _mùa_

Kiểu dữ liệu _mùa_ không sắp xếp theo thứ tự nên mình sử dụng One-hot encoding. Để ý các trường hợp có thể xảy ra trong các cột features trong data:
1. cột đấy không có đủ 4 mùa -> nếu One-hot encode như bình thường sẽ dẫn đến vector sau encode có 3 chiều -> không thống nhất được. Để xử lý trường hợp này thì mình mặc định vector sau encode phải là 4 chiều theo thự tự ý nghĩa <Xuân, Hạ, Thu, Đông>
2. một số dòng của cột _mùa_ không có dữ liệu (NaN) thì không nên encode là <0,0,0,0>. Mình đề xuất để toàn bộ là <NaN, NaN, NaN, NaN> để sau này xử lý _numeric_ sau

```python
def handle_category_data(data):
    # Định nghĩa dãy mùa cần mã hóa: Xuân, Hạ, Thu, Đông
    all_categories = ['Spring', 'Summer', 'Fall', 'Winter']
    
    # Tạo bản sao của dataframe để không thay đổi dữ liệu gốc
    df_encoded = data.copy()
    
    # Mã hóa từng cột mùa trong dataframe
    for column in df_encoded.columns:
        # Kiểm tra nếu cột là kiểu object (category cũng sẽ là object)
        if df_encoded[column].dtype == 'object' or df_encoded[column].dtype.name == 'category':
            
            # Tạo 4 cột One-Hot cho các mùa: Spring, Summer, Fall, Winter
            encoded_data = pd.DataFrame(columns=[f"{column}_{season}" for season in all_categories])
            
            # Duyệt qua từng giá trị trong cột mùa
            for idx, value in df_encoded[column].items():
                if pd.isna(value):  # Nếu là NaN, mã hóa thành NaN cho tất cả các mùa
                    encoded_data.loc[idx] = [np.nan] * len(all_categories)
                elif value in all_categories:  # Nếu là một mùa, mã hóa theo One-Hot
                    encoded_data.loc[idx] = [1 if season == value else 0 for season in all_categories]
                else:  # Nếu có giá trị ngoài mùa (hoặc không đúng mùa), mã hóa thành NaN
                    encoded_data.loc[idx] = [np.nan] * len(all_categories)

            # Thêm các cột mã hóa vào dataframe gốc và bỏ cột mùa
            df_encoded = pd.concat([df_encoded.drop(column, axis=1), encoded_data], axis=1)
    
    return df_encoded
```

Test hàm OneHot encoder trên hoạt động đúng ý không? (Mình tìm không thấy OneHotEncoder của thư viện xử lý NaN như ý nên đành tự implement)

```python
# Test hàm với một DataFrame ví dụ
data = pd.DataFrame({
    'season_1': ['Spring', 'Summer', np.nan, 'Fall'],
    'other_column': [1, 2, 3, 4]  # Cột không phải mùa
})

processed_data = handle_category_data(data)
processed_data
```

# 3. Train and test model

Ý tưởng là kết hợp LabelSpreading (1 phương pháp học bán giám sát) và XGBoost

## 3.1. Chuẩn bị dữ liệu

1. Gán nhãn -1 cho các điểm dữ liệu không có label (theo chuẩn của LabelSpreading)
2. Tách thành 2 tập: có label và không có label

```python
# X là các features, y là nhãn (sii)
X = full_dtrain.drop(columns=['sii'])
y = full_dtrain['sii']

# Tạo cột nhãn không xác định (-1) cho các điểm chưa có nhãn
y_labeled = y.copy()
y_labeled[y.isna()] = -1  # Gán giá trị -1 cho các điểm chưa có nhãn

# Tách dữ liệu có nhãn và không có nhãn
X_labeled = X[~y.isna()]  # Các điểm có nhãn
y_labeled_points = y_labeled[~y.isna()]  # Các nhãn đã biết

X_unlabeled = X[y.isna()]  # Các điểm chưa có nhãn
```

```python
labeled_data = pd.concat([X_labeled, y_labeled_points], axis=1)
labeled_data
```

## 3.2. Áp dụng LabelSpreading

```python
from sklearn.semi_supervised import LabelSpreading

# Khởi tạo mô hình LabelSpreading
label_spreader = LabelSpreading(kernel='knn', n_neighbors=5)

# Huấn luyện mô hình LabelSpreading với dữ liệu có nhãn
label_spreader.fit(X_labeled, y_labeled_points)

# Dự đoán nhãn cho các điểm chưa có nhãn
y_spread = label_spreader.predict(X_unlabeled)

# Gán nhãn cho dữ liệu chưa có nhãn
y_labeled.loc[y.isna()] = y_spread
```

Dữ liệu có được sau khi gán nhãn

```python
# Kết hợp dữ liệu đã được lan truyền nhãn
X_full = pd.concat([X_labeled, X_unlabeled])
y_full = y_labeled
```

```python
full_data = pd.concat([X_full, y_full], axis=1)
full_data
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE()
X_sm,Y_sm = sm.fit_resample(X_full,y_full)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_sm, Y_sm, test_size=0.2, random_state=42)
```

## 3.3. Huấn luyện XGBoost model

```python
from sklearn.metrics import cohen_kappa_score, make_scorer

# Hàm tính QWK cho GridSearchCV
def quadratic_kappa_score(y_true, y_pred):
    return cohen_kappa_score(y_true, y_pred, weights="quadratic")

# Tạo scorer cho GridSearchCV
qwk_scorer = make_scorer(quadratic_kappa_score, greater_is_better=True)
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV

# Định nghĩa mô hình cơ bản
xgb_model = XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss')

# Siêu tham số để tối ưu
param_grid = {
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.2],
    'n_estimators': [100, 200, 300],
    'subsample': [0.8, 1.0]
}

# GridSearchCV với QWK làm thước đo
grid_search = GridSearchCV(
    estimator=xgb_model,
    param_grid=param_grid,
    scoring=qwk_scorer,  # Sử dụng QWK
    cv=3,                # Số lần cross-validation
    verbose=1,           # Hiển thị tiến trình
    n_jobs=-1            # Sử dụng đa luồng
)

# Thực hiện tìm kiếm
grid_search.fit(X_train, y_train)

# In ra siêu tham số tốt nhất
print("Best Parameters:", grid_search.best_params_)

# Huấn luyện lại mô hình với siêu tham số tốt nhất
best_model = grid_search.best_estimator_

# Dự đoán trên tập kiểm tra
y_pred = best_model.predict(X_test)

# Tính QWK trên tập kiểm tra
qwk_score = cohen_kappa_score(y_test, y_pred, weights="quadratic")
print(f"Quadratic Weighted Kappa: {qwk_score:.4f}")
```
