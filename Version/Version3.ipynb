Since all category features are in season type whose order is not important, i prefer using One-Hot Encoding.

```python
from sklearn.preprocessing import OneHotEncoder

def handle_category_data(data):
    # Define the fixed set of seasons
    all_categories = ['Spring', 'Summer', 'Fall', 'Winter']

    df_encoded = data.copy()
    encoder = OneHotEncoder(categories=[all_categories],  # Đảm bảo 4 chiều
                             sparse_output=False,         # Đầu ra dạng mảng dày
                             handle_unknown='ignore',     # Bỏ qua giá trị ngoài danh mục
                             dtype=int)                   # Đầu ra kiểu int

    for column in data.columns:
        if df_encoded[column].dtype == 'object' or df_encoded[column].dtype.name == 'category':
            # Fit and transform the column
            encoded_array = encoder.fit_transform(df_encoded[[column]])
            
            # Generate column names
            encoded_columns = [f"{column}_{season}" for season in all_categories]
            
            # Add the encoded data to the DataFrame
            encoded_df = pd.DataFrame(encoded_array, columns=encoded_columns, index=data.index)
            df_encoded = pd.concat([df_encoded.drop(column, axis=1), encoded_df], axis=1)
    
    return df_encoded
```

**Extract and insert some important features from parquet files into data set**

```python
import os
from tqdm import tqdm
from concurrent.futures import ThreadPoolExecutor
```

```python
def process_file(filename, dirname):
    df = pd.read_parquet(os.path.join(dirname, filename, 'part-0.parquet'))
    df.drop('step', axis=1, inplace=True)
    
    return df.describe().values.reshape(-1), filename.split('=')[1]


def load_time_series(dirname):
    ids = os.listdir(dirname)
    
    with ThreadPoolExecutor() as executor:
        results = list(tqdm(executor.map(lambda fname: process_file(fname, dirname), ids), total=len(ids)))
    
    stats, indexes = zip(*results)
    df = pd.DataFrame(stats, columns=[f"stat_{i}" for i in range(len(stats[0]))])
    df['id'] = indexes
    
    return df
```

**Analysis for choosing imputer stragy**

```python
import pandas as pd
from scipy.stats import shapiro, expon, uniform, kstest

# Kiểm tra phân phối chuẩn
def test_normality(data_column):
    stat, p_value = shapiro(data_column.dropna())
    print(f"Shapiro-Wilk Test: Statistic = {stat}, P-value = {p_value}")
    return p_value > 0.05  # p-value > 0.05 nghĩa là có thể coi là phân phối chuẩn

# Kiểm tra phân phối mũ
def test_exponential(data_column):
    params = expon.fit(data_column.dropna())
    D, p_value = kstest(data_column.dropna(), 'expon', args=params)
    print(f"Exponential Distribution Test: D-statistic = {D}, P-value = {p_value}")
    return p_value > 0.05  # Nếu p-value > 0.05, dữ liệu có thể tuân theo phân phối mũ

# Kiểm tra phân phối đồng đều
def test_uniform(data_column):
    params = uniform.fit(data_column.dropna())
    D, p_value = kstest(data_column.dropna(), 'uniform', args=params)
    print(f"Uniform Distribution Test: D-statistic = {D}, P-value = {p_value}")
    return p_value > 0.05  # Nếu p-value > 0.05, dữ liệu có thể tuân theo phân phối đồng đều

# Xác định phân phối của cột dữ liệu
def determine_distribution(data_column):
    if test_normality(data_column):
        return 'Normal'
    elif test_exponential(data_column):
        return 'Exponential'
    elif test_uniform(data_column):
        return 'Uniform'
    else:
        return 'Unknown'  # Nếu không phù hợp với bất kỳ phân phối nào
```

```python
from scipy.stats import shapiro
from scipy.stats import norm, expon, uniform

# Hàm kiểm tra phân phối chuẩn cho từng cột
def check_normality(data):
    results = {}
    for column in data.columns:
        if pd.api.types.is_numeric_dtype(data[column]):
            col_data = data[column].dropna()  # Loại bỏ các giá trị thiếu cho mỗi cột
            stat, p_value = shapiro(col_data)
            results[column] = {
                'Statistic': stat,
                'P-Value': p_value,
                'Is Normal': p_value > 0.05  # Nếu p-value > 0.05, coi là phân phối chuẩn
            }
    return pd.DataFrame(results).T


# Hàm ước lượng các tham số phân phối cho từng cột
def fit_distributions(data):
    fitted_results = {}
    for column in data.columns:
        if pd.api.types.is_numeric_dtype(data[column]):
            col_data = data[column].dropna()  # Loại bỏ các giá trị thiếu cho mỗi cột
            norm_params = norm.fit(col_data)
            expon_params = expon.fit(col_data)
            uniform_params = uniform.fit(col_data)

            fitted_results[column] = {
                'Normal': norm_params,
                'Exponential': expon_params,
                'Uniform': uniform_params
            }
    return fitted_results
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')

df_imputed = pd.DataFrame(imputer.fit_transform(labeled_data_X), columns=labeled_data_X.columns)

print("Dữ liệu sau khi xử lý:")
df_imputed
```

# Train and test model

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE()
X_sm,Y_sm = sm.fit_resample(df_imputed,labeled_data_y)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_sm, Y_sm, test_size=0.2, random_state=42)
```

```python
from sklearn.metrics import cohen_kappa_score, make_scorer

# Hàm tính QWK cho GridSearchCV
def quadratic_kappa_score(y_true, y_pred):
    return cohen_kappa_score(y_true, y_pred, weights="quadratic")

# Tạo scorer cho GridSearchCV
qwk_scorer = make_scorer(quadratic_kappa_score, greater_is_better=True)
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV

# Định nghĩa mô hình cơ bản
xgb_model = XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss')

# Siêu tham số để tối ưu
param_grid = {
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.2],
    'n_estimators': [100, 200, 300],
    'subsample': [0.8, 1.0]
}

# GridSearchCV với QWK làm thước đo
grid_search = GridSearchCV(
    estimator=xgb_model,
    param_grid=param_grid,
    scoring=qwk_scorer,  # Sử dụng QWK
    cv=3,                # Số lần cross-validation
    verbose=1,           # Hiển thị tiến trình
    n_jobs=-1            # Sử dụng đa luồng
)

# Thực hiện tìm kiếm
grid_search.fit(X_train, y_train)

# In ra siêu tham số tốt nhất
print("Best Parameters:", grid_search.best_params_)

# Huấn luyện lại mô hình với siêu tham số tốt nhất
best_model = grid_search.best_estimator_

# Dự đoán trên tập kiểm tra
y_pred = best_model.predict(X_test)

# Tính QWK trên tập kiểm tra
qwk_score = cohen_kappa_score(y_test, y_pred, weights="quadratic")
print(f"Quadratic Weighted Kappa: {qwk_score:.4f}")
```
